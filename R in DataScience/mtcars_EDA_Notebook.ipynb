---
## Step 1: Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Nicer plots
sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (8, 5)
plt.rcParams["figure.dpi"] = 100
print("Libraries loaded ✅")
```

---
## Step 2: Load the Dataset

We'll load `mtcars` from a CSV URL (it's also available in many Python packages).

```python
# Load mtcars from a public URL
url = "https://gist.githubusercontent.com/seankross/a412dfbd88b3db70b74b/raw/5f23f993cd87c283ce766e7ac6b329ee7cc2e1d1/mtcars.csv"
mtcars = pd.read_csv(url)

# Set car name as index
mtcars = mtcars.set_index("model")
print(f"Dataset loaded: {mtcars.shape[0]} rows × {mtcars.shape[1]} columns")
mtcars.head()
```

---
## Step 3: Inspect the Data

```python
# Dimensions
print(f"Rows: {mtcars.shape[0]}")
print(f"Columns: {mtcars.shape[1]}")
print(f"\nColumn names: {list(mtcars.columns)}")
print(f"\nData types:")
print(mtcars.dtypes)
```

```python
# Check for missing values
print("Missing values per column:")
print(mtcars.isnull().sum())
print(f"\nTotal missing: {mtcars.isnull().sum().sum()}")
```

```python
# First and last rows
print("=== First 5 rows ===")
display(mtcars.head())

print("\n=== Last 5 rows ===")
display(mtcars.tail())
```

**Observation:** 32 rows × 11 columns. All numeric types. No missing values. Some columns (`cyl`, `vs`, `am`, `gear`, `carb`) are really categorical.

---
## Step 4: Summary Statistics

```python
# Descriptive statistics
mtcars.describe().round(2)
```

```python
# Key statistics for MPG
print("=== MPG Statistics ===")
print(f"Mean:   {mtcars['mpg'].mean():.2f}")
print(f"Median: {mtcars['mpg'].median():.2f}")
print(f"Std:    {mtcars['mpg'].std():.2f}")
print(f"Min:    {mtcars['mpg'].min():.2f}")
print(f"Max:    {mtcars['mpg'].max():.2f}")

print("\n=== HP Statistics ===")
print(f"Mean:   {mtcars['hp'].mean():.2f}")
print(f"Std:    {mtcars['hp'].std():.2f}")

print("\n=== Weight Statistics ===")
print(f"Mean:   {mtcars['wt'].mean():.2f} (× 1000 lbs)")
print(f"Std:    {mtcars['wt'].std():.2f}")
```

---
## Step 5: Frequency Counts for Categorical Variables

```python
print("Cylinders:")
print(mtcars["cyl"].value_counts().sort_index())

print("\nTransmission (0=auto, 1=manual):")
print(mtcars["am"].value_counts().sort_index())

print("\nGears:")
print(mtcars["gear"].value_counts().sort_index())

print("\nEngine (0=V, 1=straight):")
print(mtcars["vs"].value_counts().sort_index())
```

**Findings:** Most cars have 8 cylinders (14), automatic transmission (19), and 3 gears (15).

---
## Step 6: Visualizations

### 6.1 — Histogram: MPG Distribution

```python
fig, ax = plt.subplots(figsize=(8, 5))
ax.hist(mtcars["mpg"], bins=10, color="steelblue", edgecolor="white")
ax.axvline(mtcars["mpg"].mean(), color="red", linestyle="--", linewidth=2,
           label=f"Mean = {mtcars['mpg'].mean():.1f}")
ax.set_title("Distribution of MPG", fontsize=16, fontweight="bold")
ax.set_xlabel("Miles Per Gallon")
ax.set_ylabel("Frequency")
ax.legend()
plt.tight_layout()
plt.show()
```

**Interpretation:** Most cars cluster around 15–22 MPG. The distribution is slightly right-skewed with a few efficient outliers above 30 MPG.

### 6.2 — Histogram: Horsepower

```python
fig, ax = plt.subplots(figsize=(8, 5))
ax.hist(mtcars["hp"], bins=10, color="coral", edgecolor="white")
ax.axvline(mtcars["hp"].mean(), color="darkred", linestyle="--", linewidth=2,
           label=f"Mean = {mtcars['hp'].mean():.1f}")
ax.set_title("Distribution of Horsepower", fontsize=16, fontweight="bold")
ax.set_xlabel("Horsepower")
ax.set_ylabel("Frequency")
ax.legend()
plt.tight_layout()
plt.show()
```

**Interpretation:** HP is right-skewed too — most cars are 50–175 HP, with a few high-performance outliers.

### 6.3 — Boxplot: MPG by Cylinders

```python
fig, ax = plt.subplots(figsize=(8, 5))
colors = ["lightgreen", "gold", "salmon"]
bp = mtcars.boxplot(column="mpg", by="cyl", ax=ax,
                    patch_artist=True, return_type="dict")
for patch, color in zip(bp["mpg"]["boxes"], colors):
    patch.set_facecolor(color)
ax.set_title("MPG by Number of Cylinders", fontsize=16, fontweight="bold")
ax.set_xlabel("Cylinders")
ax.set_ylabel("Miles Per Gallon")
fig.suptitle("")  # remove auto-title
plt.tight_layout()
plt.show()
```

**Interpretation:** 4-cylinder cars are the most fuel-efficient, 8-cylinder the least. Clear negative relationship.

### 6.4 — Boxplot: MPG by Transmission

```python
fig, ax = plt.subplots(figsize=(7, 5))
mtcars_plot = mtcars.copy()
mtcars_plot["Transmission"] = mtcars_plot["am"].map({0: "Automatic", 1: "Manual"})
sns.boxplot(x="Transmission", y="mpg", data=mtcars_plot,
            palette=["lightblue", "lightyellow"], ax=ax)
ax.set_title("MPG by Transmission Type", fontsize=16, fontweight="bold")
ax.set_ylabel("Miles Per Gallon")
plt.tight_layout()
plt.show()
```

**Interpretation:** Manual transmission cars tend to have higher MPG than automatics in this dataset.

### 6.5 — Scatter Plot: MPG vs Weight

```python
fig, ax = plt.subplots(figsize=(8, 5))
ax.scatter(mtcars["wt"], mtcars["mpg"], c="steelblue", s=60, edgecolor="white")

# Add regression line
z = np.polyfit(mtcars["wt"], mtcars["mpg"], 1)
p = np.poly1d(z)
x_line = np.linspace(mtcars["wt"].min(), mtcars["wt"].max(), 100)
ax.plot(x_line, p(x_line), "r-", linewidth=2, label=f"y = {z[0]:.2f}x + {z[1]:.2f}")

ax.set_title("MPG vs Weight", fontsize=16, fontweight="bold")
ax.set_xlabel("Weight (× 1000 lbs)")
ax.set_ylabel("Miles Per Gallon")
ax.legend()
plt.tight_layout()
plt.show()

r = mtcars["mpg"].corr(mtcars["wt"])
print(f"Correlation (mpg vs wt): r = {r:.3f}")
```

**Interpretation:** Strong negative relationship (r ≈ −0.87). Heavier cars have significantly worse fuel economy.

### 6.6 — Scatter Plot: MPG vs Horsepower

```python
fig, ax = plt.subplots(figsize=(8, 5))
ax.scatter(mtcars["hp"], mtcars["mpg"], c="coral", s=60, edgecolor="white")

z = np.polyfit(mtcars["hp"], mtcars["mpg"], 1)
p = np.poly1d(z)
x_line = np.linspace(mtcars["hp"].min(), mtcars["hp"].max(), 100)
ax.plot(x_line, p(x_line), color="darkred", linewidth=2)

ax.set_title("MPG vs Horsepower", fontsize=16, fontweight="bold")
ax.set_xlabel("Horsepower")
ax.set_ylabel("Miles Per Gallon")
plt.tight_layout()
plt.show()

r = mtcars["mpg"].corr(mtcars["hp"])
print(f"Correlation (mpg vs hp): r = {r:.3f}")
```

**Interpretation:** Strong negative relationship (r ≈ −0.78). More powerful cars use more fuel.

### 6.7 — Bar Chart: Cars by Cylinder Count

```python
fig, ax = plt.subplots(figsize=(6, 4))
cyl_counts = mtcars["cyl"].value_counts().sort_index()
cyl_counts.plot(kind="bar", color=["lightgreen", "gold", "salmon"],
                edgecolor="gray", ax=ax)
ax.set_title("Number of Cars by Cylinder Count", fontsize=14, fontweight="bold")
ax.set_xlabel("Cylinders")
ax.set_ylabel("Count")
ax.set_xticklabels(cyl_counts.index, rotation=0)
plt.tight_layout()
plt.show()
```

---
## Step 7: Correlation Analysis

```python
# Correlation matrix for key variables
key_vars = ["mpg", "hp", "wt", "disp", "drat", "qsec"]
corr = mtcars[key_vars].corr().round(2)
print("Correlation Matrix:")
display(corr)
```

```python
# Heatmap of correlations
fig, ax = plt.subplots(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap="RdBu_r", center=0, vmin=-1, vmax=1,
            square=True, linewidths=1, ax=ax)
ax.set_title("Correlation Heatmap — Key Variables", fontsize=14, fontweight="bold")
plt.tight_layout()
plt.show()
```

**Key correlations:**
- **mpg vs wt:** −0.87 (strong negative — heavier cars use more fuel)
- **mpg vs disp:** −0.85 (strong negative — bigger engines use more fuel)
- **mpg vs hp:** −0.78 (strong negative — more HP = worse fuel economy)
- **wt vs disp:** +0.89 (strong positive — heavy cars have big engines)

---
## Step 8: Pairplot — Multiple Relationships

```python
# Pairplot for key variables
sns.pairplot(mtcars[["mpg", "hp", "wt", "disp"]],
             plot_kws={"color": "steelblue", "s": 30, "edgecolor": "white"},
             diag_kws={"color": "steelblue", "edgecolor": "white"})
plt.suptitle("Pairplot — Key Variables", y=1.02, fontsize=14, fontweight="bold")
plt.tight_layout()
plt.show()
```

**What to look for:** Strong linear trends (mpg vs wt), cloud-like scatter (weak relationship), curved patterns (non-linear).

---
## Step 9: Subsetting and Aggregation

```python
# Filter: fuel-efficient cars (MPG > 20)
efficient = mtcars[mtcars["mpg"] > 20][["mpg", "cyl", "hp", "wt", "am"]]
print(f"Fuel-efficient cars (MPG > 20): {len(efficient)} out of 32\n")
display(efficient)
```

```python
# Mean MPG by cylinder group
print("Mean MPG by Cylinders:")
print(mtcars.groupby("cyl")["mpg"].mean().round(2))

print("\nMean MPG by Transmission:")
am_means = mtcars.groupby("am")["mpg"].mean().round(2)
am_means.index = ["Automatic", "Manual"]
print(am_means)

print("\nMean MPG by Gears:")
print(mtcars.groupby("gear")["mpg"].mean().round(2))
```

```python
# Advanced filtering: High-efficiency manual cars
high_eff = mtcars[(mtcars["mpg"] > 25) & (mtcars["am"] == 1)][["mpg","cyl","hp","wt","gear"]]
print(f"High-efficiency manual cars (MPG > 25): {len(high_eff)}\n")
display(high_eff)
```

```python
# 4-cylinder cars sorted by MPG
four_cyl = (mtcars[mtcars["cyl"] == 4][["mpg","hp","wt","gear","am"]]
            .sort_values("mpg", ascending=False))
print("4-cylinder cars (sorted by MPG):")
display(four_cyl)
```

## EDA Cheat Sheet — Python

| Task | Python Code |
|:-----|:------------|
| View data | `df.head()`, `df.tail()` |
| Dimensions | `df.shape` |
| Data types | `df.dtypes` |
| Missing values | `df.isnull().sum()` |
| Summary | `df.describe()` |
| Frequency | `df['col'].value_counts()` |
| Histogram | `plt.hist()` or `df['col'].hist()` |
| Boxplot | `sns.boxplot()` or `df.boxplot()` |
| Scatter | `plt.scatter()` or `sns.scatterplot()` |
| Correlation | `df.corr()` |
| Heatmap | `sns.heatmap(df.corr())` |
| Filter | `df[df['col'] > value]` |
| Group by | `df.groupby('col').mean()` |
