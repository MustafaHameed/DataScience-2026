## What is Association Rule Mining?

**Association Rule Mining (ARM)** is a data mining technique that discovers *interesting relationships* (associations) among items in large datasets.

**Classic application:** Market Basket Analysis — which products do customers frequently purchase together?

> *"Customers who buy a laptop and a mouse also tend to buy a keyboard."*  
> `{laptop, mouse}` → `{keyboard}`

### Key Definitions

| Term | Definition |
|------|-----------|
| **Transaction** | A single purchase event (one receipt / one basket) |
| **Item** | An individual product in a transaction |
| **Itemset** | A collection of one or more items, e.g. `{laptop, mouse}` |
| **Association Rule** | An implication: `{A} → {B}` ("if A then B") |
| **Antecedent (LHS)** | The "if" part — `{laptop, mouse}` |
| **Consequent (RHS)** | The "then" part — `{keyboard}` |

### Why Does ARM Matter?

| Business Uses | Other Domains |
|--------------|---------------|
| Product placement — put associated items nearby | Medical co-diagnoses |
| Bundle deals — "buy laptop + mouse, get keyboard 20% off" | Web click-stream patterns |
| Cross-selling — recommend items at checkout | Course enrollment patterns |
| Inventory — stock associated items together | Fraud detection |

---

## The Four Key Measures of ARM

Every association rule is evaluated by **four** important measures:

### Measure 1 — Support
> **Definition:** Support measures how *frequently* an itemset appears in the dataset. It tells us how **common** the pattern is.

$$\text{Support}(X) = \frac{\text{Number of transactions containing } X}{\text{Total number of transactions } (N)}$$

- Higher support = more common pattern
- Used to **filter out rare patterns** (set a minimum threshold)

### Measure 2 — Confidence
> **Definition:** Confidence measures how *reliable* the rule is. Given that the antecedent (LHS) is purchased, how often is the consequent (RHS) also purchased?

$$\text{Confidence}(A \Rightarrow B) = \frac{\text{Support}(A \cup B)}{\text{Support}(A)}$$

- Ranges from 0 to 1 (0% to 100%)
- ⚠️ High confidence alone can be misleading if B is very popular!

### Measure 3 — Lift
> **Definition:** Lift measures whether A and B appear together *more often than expected by chance*.

$$\text{Lift}(A \Rightarrow B) = \frac{\text{Confidence}(A \Rightarrow B)}{\text{Support}(B)}$$

| Lift Value | Meaning |
|-----------|---------|
| **Lift > 1** | Positive association — items appear together more than expected |
| **Lift = 1** | No association — items are independent |
| **Lift < 1** | Negative association — items *avoid* each other |

### Measure 4 — Conviction
> **Definition:** Conviction measures the degree of **dependence** between antecedent and consequent.

$$\text{Conviction}(A \Rightarrow B) = \frac{1 - \text{Support}(B)}{1 - \text{Confidence}(A \Rightarrow B)}$$

| Conviction | Meaning |
|-----------|---------|
| **Conv = 1** | Items are independent |
| **Conv > 1** | Rule is meaningful — items are dependent |
| **Conv → ∞** | Perfect rule (confidence = 100%) |

## Hand Calculation — Rule 1: {laptop, mouse} → {keyboard}

### Step 1 — Support
Transactions containing {laptop, mouse, keyboard}: **T1, T5, T7, T9** → Count = **4**

$$\text{Support}(\{laptop, mouse, keyboard\}) = \frac{4}{10} = 0.40$$

✅ **40% of all transactions** contain laptop + mouse + keyboard together.

### Step 2 — Confidence
Transactions with {laptop, mouse}: **T1, T2, T5, T7, T9** → Count = **5**, Support = 5/10 = **0.50**

$$\text{Confidence} = \frac{0.40}{0.50} = 0.80$$

✅ When a customer buys laptop + mouse, there is an **80% chance** they also buy a keyboard.

### Step 3 — Lift
Support({keyboard}) = 8/10 = **0.80**

$$\text{Lift} = \frac{0.80}{0.80} = 1.00$$

⚠️ Lift = 1.00 → laptop + mouse does **not** increase or decrease the chance of buying keyboard — they are **independent**.

### Step 4 — Conviction

$$\text{Conviction} = \frac{1 - 0.80}{1 - 0.80} = \frac{0.20}{0.20} = 1.00$$

⚠️ Conviction = 1.00 confirms **independence**.

> **Key Lesson:** High confidence (0.80) but Lift = 1.0! The keyboard is in 80% of transactions, so high confidence is misleading. **Always check Lift!**

## Hand Calculation — Rule 2: {mouse} → {mousepad}

| Step | Calculation | Result |
|------|------------|--------|
| Txns with {mouse, mousepad} | T4, T6, T9 | **3** |
| Support({mouse, mousepad}) | 3/10 | **0.30** |
| Support({mouse}) | 7/10 | **0.70** |
| Support({mousepad}) | 3/10 | **0.30** |
| **Confidence** | 0.30 / 0.70 | **0.429** |
| **Lift** | 0.429 / 0.30 | **1.43** |
| **Conviction** | (1−0.30) / (1−0.429) | **1.23** |

✅ **Lift = 1.43** — Buying a mouse increases the chance of buying a mousepad by 43% over random. This is a **useful rule**!

## Hand Calculation — Rule 3: {laptop} → {mouse}

| Step | Calculation | Result |
|------|------------|--------|
| Txns with {laptop, mouse} | T1,T2,T5,T7,T9 | **5** |

| Support({laptop}) | 7/10 | **0.70** |
| Support({mouse}) | 7/10 | **0.70** |
| **Confidence** | 0.50 / 0.70 | **0.714** |
| **Lift** | 0.714 / 0.70 | **1.02** |
| **Conviction** | (1−0.70)/(1−0.714) | **1.05** |

Lift = 1.02, Conviction = 1.05 — very *slight* positive association. Nearly independent.

## Step 4 — Verify Hand Calculations in R

Let's confirm our by-hand calculations match what R computes.

```python
# Step 4a — Verify Rule 1: {laptop, mouse} -> {keyboard}
cat("=== Rule 1: {laptop, mouse} -> {keyboard} ===\n\n")

supp_lmk <- sum(sapply(baskets, function(b) all(c("laptop","mouse","keyboard") %in% b))) / 10
supp_lm  <- sum(sapply(baskets, function(b) all(c("laptop","mouse") %in% b))) / 10
supp_k   <- sum(sapply(baskets, function(b) "keyboard" %in% b)) / 10

conf1 <- supp_lmk / supp_lm
lift1 <- conf1 / supp_k
conv1 <- (1 - supp_k) / (1 - conf1)

cat("Support({laptop,mouse,keyboard}) =", supp_lmk, "\n")
cat("Support({laptop,mouse})          =", supp_lm,  "\n")
cat("Support({keyboard})              =", supp_k,   "\n\n")
cat("Confidence =", round(conf1, 4), "\n")
cat("Lift       =", round(lift1, 4), "\n")
cat("Conviction =", round(conv1, 4), "\n")
```

```python
# Step 4b — Verify Rule 2: {mouse} -> {mousepad}
cat("=== Rule 2: {mouse} -> {mousepad} ===\n\n")

supp_mm  <- sum(sapply(baskets, function(b) all(c("mouse","mousepad") %in% b))) / 10
supp_m   <- sum(sapply(baskets, function(b) "mouse" %in% b)) / 10
supp_mp  <- sum(sapply(baskets, function(b) "mousepad" %in% b)) / 10

conf2 <- supp_mm / supp_m
lift2 <- conf2 / supp_mp
conv2 <- (1 - supp_mp) / (1 - conf2)

cat("Support({mouse,mousepad}) =", supp_mm, "\n")
cat("Support({mouse})          =", supp_m,  "\n")
cat("Support({mousepad})       =", supp_mp, "\n\n")
cat("Confidence =", round(conf2, 4), "\n")
cat("Lift       =", round(lift2, 4), "\n")
cat("Conviction =", round(conv2, 4), "\n")
```

```python
# Step 4c — Verify Rule 3: {laptop} -> {mouse}
cat("=== Rule 3: {laptop} -> {mouse} ===\n\n")

supp_lm2 <- sum(sapply(baskets, function(b) all(c("laptop","mouse") %in% b))) / 10
supp_l   <- sum(sapply(baskets, function(b) "laptop" %in% b)) / 10
supp_m2  <- sum(sapply(baskets, function(b) "mouse" %in% b)) / 10

conf3 <- supp_lm2 / supp_l
lift3 <- conf3 / supp_m2
conv3 <- (1 - supp_m2) / (1 - conf3)

cat("Support({laptop,mouse}) =", supp_lm2, "\n")
cat("Support({laptop})       =", supp_l,   "\n")
cat("Support({mouse})        =", supp_m2,  "\n\n")
cat("Confidence =", round(conf3, 4), "\n")
cat("Lift       =", round(lift3, 4), "\n")
cat("Conviction =", round(conv3, 4), "\n")
```

## Step 5 — Find Frequent Itemsets with Apriori

The **Apriori algorithm** works in two stages:
1. Find all itemsets with support ≥ min_support (**frequent itemsets**)
2. Generate rules from those itemsets with confidence ≥ min_confidence

**Anti-monotone property:** A subset of a frequent itemset must also be frequent. This allows pruning!

```python
# Step 5 — Find Frequent Itemsets (min support = 20%)
frequent_sets <- apriori(
  txns,
  parameter = list(supp = 0.20, target = "frequent itemsets")
)

cat("Frequent itemsets found:", length(frequent_sets), "\n\n")
inspect(sort(frequent_sets, by = "support", decreasing = TRUE))
```

## Step 6 — Generate Association Rules

```python
# Step 6 — Generate Association Rules (min support=20%, min confidence=50%)
rules <- apriori(
  txns,
  parameter = list(supp = 0.20, conf = 0.50, minlen = 2)
)

cat("Rules generated:", length(rules), "\n\n")

# Sort by lift (most interesting first)
rules_sorted <- sort(rules, by = "lift", decreasing = TRUE)
inspect(rules_sorted)
```

## Step 7 — How to Read the R Output

| Column | Meaning |
|--------|---------|
| **lhs** | Left-hand side (antecedent) — the "if" items |
| **rhs** | Right-hand side (consequent) — the "then" item |
| **support** | Fraction of all transactions containing both sides |
| **confidence** | Of transactions with lhs, what fraction also has rhs? |
| **coverage** | Support of lhs alone |
| **lift** | > 1 = positive · = 1 = independent · < 1 = negative |
| **count** | Absolute number of matching transactions |

**Example:** `{laptop, mouse} → {keyboard} [supp=0.40, conf=0.80, lift=1.00]`  
*"40% of transactions have all three. When laptop + mouse are bought, 80% also buy keyboard. But lift=1 — no better than chance because keyboard is already very popular (80% support)."*

## Step 9 — Filter Rules for a Specific Item

In practice, you often want rules for one specific product.

```python
# Step 9 — What items predict buying a keyboard?
rules_keyboard <- apriori(
  txns,
  parameter  = list(supp = 0.15, conf = 0.50, minlen = 2),
  appearance = list(rhs = "keyboard", default = "lhs")
)

cat("Rules predicting keyboard:", length(rules_keyboard), "\n\n")
inspect(sort(rules_keyboard, by = "lift", decreasing = TRUE))
```

## How to Interpret Results — Practical Guide

| Measure | What to look for |
|---------|-----------------|
| **Support** | Filter out rare patterns (min 0.10–0.30) |
| **Confidence** | High confidence (≥ 0.50) = rule fires correctly often |
| **Lift** | **> 1** = useful · **> 1.5** = strong · **> 2** = very strong |
| **Conviction** | **> 1** = dependent · **> 1.5** = strong dependency |

### Business Actions Based on Results

| Lift Range | Action |
|-----------|--------|
| Lift > 2.0 | **Strong bundle** — create a package deal |
| Lift 1.5–2.0 | **Cross-sell** — recommend at checkout |
| Lift 1.1–1.5 | **Place nearby** — put on adjacent shelves |
| Lift ≈ 1.0 | **No action** — items are independent |
| Lift < 1.0 | **Separate** — avoid placing together |

### ⚠️ Common Pitfall
**High confidence + low lift** → the consequent is just very popular!  
Example: Confidence({laptop,mouse}→{keyboard}) = 0.80, but Lift = 1.0 because keyboard is in 80% of transactions.  
**Always check Lift and Conviction before acting on a rule!**

## Summary

| Measure | Formula | Good Values |
|---------|---------|-------------|
| **Support** | freq(A∪B) / N | ≥ 0.10 |
| **Confidence** | Supp(A∪B) / Supp(A) | ≥ 0.50 |
| **Lift** | Conf / Supp(B) | **> 1** (higher = better) |
| **Conviction** | (1−Supp(B)) / (1−Conf) | **> 1** (higher = stronger) |

### Key Takeaways
- **Support** = how common is the pattern?
- **Confidence** = how reliable is the rule?
- **Lift > 1** = better than random guessing
- **Conviction > 1** = items are genuinely dependent
- Always check **Lift** before trusting a high-confidence rule
- The **Apriori algorithm** prunes efficiently using the anti-monotone property
